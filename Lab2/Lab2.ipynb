```python
from thinkdsp import Sinusoid
from thinkdsp import normalize, unbias
import numpy as np

class SawtoothSignal(Sinusoid):
    def evaluate(self, ts):
        cycles = self.freq * ts + self.offset / np.pi / 2
        frac, _ = np.modf(cycles)
        ys = normalize(unbias(frac), self.amp)
        return ys
```

```python
sawtooth = SawtoothSignal()
sawtooth = sawtooth.make_wave(1, 30000)
sawtooth.make_audio()
```

```python
sawtooth.make_spectrum().plot()
```

```python
def del_spectrum(spectrum):
    spectrum.hs[1:] /= spectrum.fs[1:]
    spectrum.hs[0] = 0
```

```python
wave = SquareSignal(460).make_wave(0.5)
wave.make_audio()
```

```python
spectrum = wave.make_spectrum()
spectrum.plot()
```

```python
del_spectrum(spectrum)
spectrum.scale(460)
spectrum.plot()
```

```python
new = spectrum.make_wave()
new.make_audio()
```
